```python
def unpack_json(input_text):
    """
    Extract the JSON part of the input text starting from the first '{'.
    Parses the JSON if possible, otherwise returns the extracted string or error message.
    """
    try:
        # Ensure the input is a valid string
        if not isinstance(input_text, (str, bytes, bytearray)):
            raise ValueError("Input must be a string, bytes, or bytearray.")

        # Find the index of the first '{'
        start_index = input_text.find('{')
        if start_index != -1:
            # Extract text starting from the first '{'
            json_part = input_text[start_index:]
            # Try to parse the JSON
            try:
                return json.loads(json_part)  # Return parsed JSON if valid
            except json.JSONDecodeError as e:
                return f"Extracted text is not valid JSON: {json_part}. Error: {e}"
        else:
            return "No '{' found in the input text."
    except Exception as e:
        return f"An error occurred: {e}"



def cal_score(llm_data,total_sen):
  # llm_data = orjson.loads(json_str)
  all_relevant_sentence_keys_R = llm_data["all_relevant_sentence_keys"]
  all_utilized_sentence_keys_U = llm_data["all_utilized_sentence_keys"]
  common_count = len(set(all_relevant_sentence_keys_R) & set(all_utilized_sentence_keys_U))
  relevent_score = float(len(all_relevant_sentence_keys_R))/float(total_sen) if total_sen else 0
  utilization_score = float(len(all_utilized_sentence_keys_U))/float(total_sen) if total_sen else 0
  complete_score = float(common_count)/float(len(all_relevant_sentence_keys_R)) if  relevent_score else 0
  return relevent_score,utilization_score,complete_score,llm_data["overall_supported"]


def save_file(file_name, evaluation):
    if os.path.exists(file_name):
        # If the file exists, load the existing data and append the new evaluation
        with open(file_name, "r") as f:
            existing_data = json.load(f)
        existing_data.append(evaluation)
        with open(file_name, "w") as f:
            json.dump(existing_data, f, indent=4)
    else:
        # If the file does not exist, create a new file and write the evaluation
        with open(file_name, "w") as f:
            json.dump([evaluation], f, indent=4)

def formattop_k(top_K_documents):
  output = []  # List to store the final output
  count_sent=[]
  for chunk_index, chunk in enumerate(top_K_documents):
        sentences = sent_tokenize(chunk)  # Tokenize sentences in the chunk
        chunk_sentences = []  # To store sentences of the current chunk

        sentence_counter = 0  # Initialize the sentence counter
        for sentence_index, sentence in enumerate(sentences, start=1):
            if len(sentence) > 1:  # Check if sentence length is greater than 1
                sentence_counter += 1  # Increment the counter only for valid sentences
                # Convert index to a corresponding lowercase letter (a, b, c, ...)
                sentence_label = f"{chunk_index}{string.ascii_lowercase[sentence_counter - 1]}"
                chunk_sentences.append([sentence_label, sentence])  # Add sentence to the chunk list
                count_sent.append([sentence_label, sentence])
        if chunk_sentences:
            output.append(chunk_sentences)  # Add the list of sentences for the current chunk to the output

  # print(len(count_sent))
  return output,len(count_sent)

def cal_cosine_sim(toembed_text1,toembed_text2,emd_obj):
  response_emd1=emd_obj.embed_query(toembed_text1)
  response_emd2=emd_obj.embed_query(toembed_text2)
  # response_emd1=hf_embed_model.encode(toembed_text1, normalize_embeddings=True)
  # response_emd2=hf_embed_model.encode(toembed_text2, normalize_embeddings=True)
  return cosine_similarity([response_emd1], [response_emd2])[0][0]

def generate_embedding(toembed_text,emd_obj):
    emd = emd_obj.embed_query(toembed_text)
    return emd
  # Generate embeddings

def save_file(file_path, dump_text):
    # if os.path.exists(file_path):
    #     os.remove(file_path)  # Deletes the file
    #     print(f"{file_path} has been deleted.")

    # Create or overwrite the file with the new JSON content
    with open(file_path, "w") as f:
        json.dump(dump_text, f, indent=4)
        print(f"Added to file")

def cal_rmse(prediction,actual):
  return np.sqrt(mean_squared_error(prediction,actual))
def cal_auc_score(prediction,actual):
  return roc_auc_score(prediction,actual)
```

```python
def load_and_combine_datasets(dataset_names):
    all_dataframes = []

    for dataset in dataset_names:
        try:
            # Load the dataset
            data = load_dataset("rungalileo/ragbench", dataset, split="test")

            # Convert to a Pandas DataFrame
            df = pd.DataFrame(data)

            # Add a column to indicate the source dataset
            df['dataset_name'] = dataset

            # Append to the list
            all_dataframes.append(df)
            print(f"Successfully loaded dataset: {dataset}")
        except Exception as e:
            print(f"Failed to load dataset: {dataset}. Error: {e}")

    # Combine all DataFrames into a single DataFrame
    if all_dataframes:
        combined_df = pd.concat(all_dataframes, ignore_index=True)
        return combined_df
    else:
        print("No datasets were successfully loaded.")
        return pd.DataFrame()
```

```python
def create_chunks_from_dataframe(dataframe, splitter):
    """
    Iterate over a DataFrame, split text into chunks, and store both the chunks and their metadata.

    Args:
        dataframe (pd.DataFrame): The DataFrame containing the data to be chunked.
            Expected columns:
                - "documents": The text or list of texts to split.
                - "id": A unique identifier for each row.
        splitter: An instance of a text splitter class with a `create_documents` method.

    Returns:
        tuple: Two lists:
            - chunked_data (list): List of chunked documents with metadata.
            - chunked_data_pagedata (list of dict): List of dictionaries containing ID and chunk content.
    """
    chunked_data = []

    for _, row in dataframe.iterrows():
        # print(f"Executing row with ID: {row['id']}")
        text_to_split = row["documents"]

        # Handle case where "documents" is a list
        if isinstance(text_to_split, list):
            text_to_split = " ".join(text_to_split)

        # Create chunks using the splitter
        chunks = splitter.create_documents(
            [text_to_split],  # Text to split
            metadatas=[{"id": row["id"]}]  # Attach metadata
        )

        # Store chunks and metadata
        for chunk in chunks:
            # chunked_data_pagedata.append({"id": row["id"], "chunk": chunk.page_content})
            chunked_data.append(chunk)

    return chunked_data
```

```python
def manage_chroma_collections(chroma_client, collection_name, embedding_function, delete_existing=True):
    """
    Delete existing collections (if specified) and create or retrieve a Chroma collection.

    Args:
        chroma_client: An instance of Chroma client (e.g., PersistentClient).
        collection_name (str): The name of the collection to create or retrieve.
        embedding_function: The embedding function to be used with the Chroma collection.
        delete_existing (bool): Whether to delete all existing collections before creating a new one.

    Returns:
        object: A Chroma collection object.
    """
    collections = chroma_client.list_collections()

    if collection_name in collections:
        if delete_existing:
            # Delete the existing collection
            chroma_client.delete_collection(collection_name)
            print(f"Deleted existing collection: {collection_name}")
        else:
            print(f"Collection '{collection_name}' already exists and delete_existing=False. No action taken.")
            return chroma_client.get_collection(collection_name)

    # Create or get the specified collection
    try:
        collection = chroma_client.get_or_create_collection(collection_name)
        vector_db = Chroma(client=chroma_client, collection_name=collection_name, embedding_function=embedding_function)
        print(f"Collection '{collection_name}' is ready.")
        return vector_db,collection
    except Exception as e:
        print(f"Error managing collection '{collection_name}': {e}")
        return None
```

```python
def process_and_add_batches(data, collection, batch_size, embedding_function):
    """
    Process data in batches, generate embeddings, and add them to a Chroma DB collection.

    Args:
        data (list): List of documents (e.g., chunked data with `page_content` attribute).
        collection: Chroma collection where documents will be added.
        batch_size (int): Number of documents to process in each batch.
        embedding_function (callable): Function to generate embeddings from text.

    Returns:a
        None
    """
    for i in range(0, len(data), batch_size):
        batch = data[i:i + batch_size]
        documents, uuids, embeddings = [], [], []

        for doc in batch:
            try:
                text = doc.page_content
                uuid = str(uuid4())
                embedding = embedding_function(text)

                documents.append(text)
                uuids.append(uuid)
                embeddings.append(embedding)
            except Exception as e:
                print(f"Error processing document: {doc}. Skipping. Error: {e}")

        try:
            # Add batch to Chroma DB
            collection.add(documents=documents, ids=uuids, embeddings=embeddings)
            print(f"Batch {i // batch_size + 1} added successfully.")
        except Exception as e:
            print(f"Error adding batch {i // batch_size + 1} to the collection. Error: {e}")
```

```python
def extract_json(json_string):
  pattern = r'^(.*?"all_utilized_sentence_keys":\s*\[.*?\]\s*})'

  # Search for the pattern
  match = re.search(pattern, json_string, re.DOTALL)
  if match:
      extracted_text = match.group(1)  # Extract the matched content
      return extracted_text
  else:
      return None
```

```python
def get_hugging_face_inference_client(documents, question, response_model, temp, api_key):
    # Set up the Inference Client
    client = InferenceClient(model=response_model, token=api_key)

    # Prepare input prompt
    input_text = f"Use the following pieces of context to answer the question Context: {documents}\n\nQuestion: {question}"

    try:
        # Generate response using text_generation method
        response = client.text_generation(input_text, temperature=temp, max_new_tokens=1024)
        return response
    except Exception as e:
        print(f"Error invoking Response LLM retrieval chain: {e}")
        return None
```

```python
def extract_json(json_string):
  pattern = r'^(.*?"all_utilized_sentence_keys":\s*\[.*?\]\s*})'

  # Search for the pattern
  match = re.search(pattern, json_string, re.DOTALL)
  if match:
      extracted_text = match.group(1)  # Extract the matched content
      return extracted_text
  else:
      return None
```

```python
def fetch_and_process_row(row_number, response_model, embedding_model):
    try:
        row_number = int(str(row_number).strip())  # Ensure input is valid
        if not (1 <= row_number <= len(df)):
            return f"Invalid row number! Enter a number between 1 and {len(df)}.", "", "", "", "", "", "", "", "", "", "", "", "", "", ""

        row = df.iloc[row_number - 1]  # Fetch the full row

        question = row.get("question", "N/A")
        ragbench_relevance = row.get("relevance_score", 0)
        ragbench_utilization = row.get("utilization_score", 0)
        ragbench_completeness = row.get("completeness_score", 0)
        ragbench_adherence = row.get("adherence_score", 0)
        print(row.get("response"))


        print(f"Processing row: {row.to_dict()}")  # Debugging

        eval_model = "deepseek-ai/DeepSeek-R1-Distill-Qwen-32B"  # Fixed eval model

        if embedding_model == 'LLM-Embedder':
            hf_embed_model_large = HuggingFaceEmbeddings(model_name="BAAI/LLM-Embedder")
            vector_db = vector_db_bge
        else:
            hf_embed_model_large = HuggingFaceEmbeddings(model_name="BAAI/bge-large-en")
            vector_db = vector_db_llm

        if response_model == 'Mistral-7B-Instruct-v0.3':
            response_model_name = "mistralai/Mistral-7B-Instruct-v0.3"
        else:
            response_model_name = "google/gemma-2-27b-it"

        # Ensure process_row_gradio is defined
        rel_score, util_score, compltness_score, adherence_score = 0, 0, 0, 0  # Default values

        try:
            # print("DEBUG: Is function defined?", 'process_row_gradio_new' in globals())
            llm_eval_response, total_sen, llm_response = process_row_gradio_new(question, chunked_data, eval_model, response_model_name, vector_db)
            print(llm_eval_response)
            if llm_eval_response is not None and "all_utilized_sentence_keys" in llm_eval_response:
                print("LLM Response Evaluating")
                extracted_json_str = extract_json(llm_eval_response)
                unpacked_json = unpack_json(extracted_json_str)
                rel_score, util_score, compltness_score, adherence_score = cal_score(unpacked_json, total_sen)
                print(rel_score,util_score, compltness_score, adherence_score)

        except NameError:
            llm_eval_response, total_sen, llm_response = None, 0, "Function process_row_gradio not defined!"

        cosine_sim = cal_cosine_sim(llm_response, row.get('response', ''), hf_embed_model_large)
        print(f"Cosine Similarity ----{cosine_sim}")
        cosine_similarity_score = cosine_sim

        # Calculate absolute differences
        diff_relevance = abs(float(ragbench_relevance) - float(rel_score))
        diff_utilization = abs(float(ragbench_utilization) - float(util_score))
        diff_completeness = abs(float(ragbench_completeness) - float(compltness_score))
        diff_adherence = "NA"

        return (
            question,
            ragbench_relevance, ragbench_utilization, ragbench_completeness, ragbench_adherence,
            rel_score, util_score, compltness_score, adherence_score,
            diff_relevance, diff_utilization, diff_completeness, diff_adherence,
            llm_response, cosine_similarity_score
        )
    except Exception as e:
        return f"Error: {str(e)}", "", "", "", "", "", "", "", "", "", "", "", "", "", ""
```
